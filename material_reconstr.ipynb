```python
import pandas as pd
```

```python
df = pd.read_excel('../pet_project/data/ass_sap.XLSX', header=2)
```

```python
df.head()
```

```python
df['Продукция КИС'].fillna(0)
df = df.drop([0,1], axis=0)
```

```python
def material_splitter(data):
    if type(data) == int:
        pass
    else:
        data = data.replace(' ', '').split(',')
    return data
```

```python
df['Продукция КИС'] = df['Продукция КИС'].apply(material_splitter)
```

```python
df.head()
```

```python
df = df.explode(column='Продукция КИС')
```

```python
material_dict_raw = df[['Код продукции(статьи) АССОИ','Продукция КИС']]
```

```python
material_dict_raw.shape
```

```python
material_dict = material_dict_raw.drop_duplicates()
```

```python
material_dict.shape
```

```python
material_dict.head()
```

```python
material_dict['Продукция КИС'] = material_dict['Продукция КИС'].astype('str')
```

```python
with pd.ExcelWriter('../pet_project/data/material_dict.xlsx') as writer:
    material_dict.to_excel(writer, sheet_name='material_row', index=False)
```
