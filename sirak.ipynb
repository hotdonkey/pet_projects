```python
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
i = 4567
empty = [i]
step = 0

while i != 1 and step < 4000:
    if i%2 == 0:
        i /= 2
        empty.append(i)
        step += 1
    else:
        i = i*3 + 1
        empty.append(i)
        step +=1

display(empty[::-1])
sns.lineplot(empty)
```
