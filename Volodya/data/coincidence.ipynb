```python
import pandas as pd
import re
```

```python
data = pd.read_excel('./coincidence.xlsx', header=None)
data.rename(columns={0:'court', 1:'full_name'}, inplace=True)
```

```python
data.head()
```

```python
data['court'] = data['court'].str.lower()
data['book'] = data['full_name'].str.split('.')
data['book'].fillna('s', inplace=True)
```

```python
temp = []
for i in data['book']:
    temp.append(i[-1].strip().lower())

data['book_short'] = temp
```

```python
temp = []
for i in data['book']:
    temp.append(i[0].strip().lower())

data['book_short_name'] = temp
```

```python
data.head()
```

```python
data['book_short'] = data['book_short'].str.replace('(','')
data['book_short'] = data['book_short'].str.replace(')','')
data['book_short'] = data['book_short'].str.replace('"','')
data['book_short'] = data['book_short'].str.replace("'","")
data['book_short'] = data['book_short'].str.replace("[","")
data['book_short'] = data['book_short'].str.replace("]","")
```

```python
for i in data['book_short']:
    if data['court'].str.contains(i, case=False).sum() and len(i.split())>2:
        print(i)
        print(data[data['book_short'] == i].index)
```

```python
for i in data['book_short_name']:
    if data['court'].str.contains(i, case=False).sum():
        print(i)
        print(data[data['book_short'] == i].index)
```
